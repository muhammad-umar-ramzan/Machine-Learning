```python
import sklearn.datasets as datasets
import pandas as pd
```

```python
iris=datasets.load_iris()
```

```python
dir(iris)
```

```python
iris_data=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
iris_data["target"]=iris["target"]
```

```python
iris_data["target_name"]=iris_data["target"].apply(lambda x: iris.target_names[x])
iris.target_names
```

```python
iris_data
```

```python
setosa=iris_data[iris_data["target"]==0]
versicolor=iris_data[iris_data["target"]==1]
virginica=iris_data[iris_data["target"]==2]
```

```python
import matplotlib.pyplot as plt
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.scatter(setosa["petal length (cm)"],setosa["petal width (cm)"],marker="+",color="blue")
plt.scatter(versicolor["petal length (cm)"],versicolor["petal width (cm)"],marker=".",color="green")
plt.scatter(virginica["petal length (cm)"],virginica["petal width (cm)"],marker="+",color="yellow")
plt.scatter(setosa["sepal length (cm)"],setosa["sepal width (cm)"],marker="+",color="black")
plt.scatter(versicolor["sepal length (cm)"],versicolor["sepal width (cm)"],marker=".",color="pink")
plt.scatter(virginica["sepal length (cm)"],virginica["sepal width (cm)"],marker="+",color="red")
plt.plot()
```

```python
x=iris_data.drop(["target","target_name"],axis=1)
y=iris_data.target_name
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
svm=SVC(C=2)
```

```python
svm.fit(x_train,y_train)
```

```python
svm.predict(x_test)
```

```python
svm.score(x_test,y_test)
```

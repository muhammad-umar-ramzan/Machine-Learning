```python
import pandas as pd
df=pd.read_csv("carprices.csv")
df.head()
```

```python
dummy=pd.get_dummies(df["Car Model"])
```

```python
dummy
```

```python
dummy.replace([True,False],[1,0],inplace=True)
```

```python
dummy
```

```python
final=pd.concat([dummy,df],axis=1)
```

```python
final
```

```python
final.drop(["Audi A5","Car Model"],axis=1,inplace=True)
```

```python
final
```

```python
x=final.drop(["Sell Price($)"],axis=True)
x
```

```python
y=final["Sell Price($)"]
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.predict([[0,1,45000,4]])
```

```python
reg.predict([[1,0,86000,7]])
```

```python
round(reg.score(x,y)*100,2)
```

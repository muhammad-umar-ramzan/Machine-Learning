```python
import pandas as pd
```

```python
df=pd.read_csv("Melbourne_housing_FULL.csv")
```

```python
df=df.drop(["Address","Date","Postcode","YearBuilt","Lattitude","Longtitude"],axis=1)
```

```python
df.head()
```

```python
df.nunique()
```

```python
df.isna().sum()
```

```python
col=["Propertycount","Bedroom2","Bathroom","Car","Distance"]
```

```python
df[col]=df[col].fillna(0)
```

```python
df.isna().sum()
```

```python
df['Landsize']=df['Landsize'].fillna(df['Landsize'].mean())
df['BuildingArea']=df['BuildingArea'].fillna(df['BuildingArea'].mean())
```

```python
df.isna().sum()
```

```python
df=df.dropna()
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
ce=LabelEncoder()
df['Suburb']=ce.fit_transform(df['Suburb'])
df['Type']=ce.fit_transform(df['Type'])
df['SellerG']=ce.fit_transform(df['SellerG'])
df['Method']=ce.fit_transform(df['Method'])
df['CouncilArea']=ce.fit_transform(df['CouncilArea'])
df['Regionname']=ce.fit_transform(df['Regionname'])

```

```python
x=df.drop(["Price"],axis=1)
y=df.Price
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
reg.score(x_test,y_test)
```

```python
reg.score(x_train,y_train)
```

```python
from sklearn.linear_model import Lasso
a=Lasso(alpha=50,max_iter=100,tol=0.1)
a.fit(x_train,y_train)
a.score(x_test,y_test)
a.score(x_train,y_train)
```

```python
from sklearn import linear_model
ridge=linear_model.Ridge(alpha=50,max_iter=100,tol=0.1)
ridge.fit(x_train,y_train)
```

```python
ridge.score(x_test,y_test)
```

```python
ridge.score(x_train,y_train)
```
